```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def euler(volt_func, volt_init, n_func, c_func, n_init, c_init, current_funcs, t_range, dt):

    t = np.arange(t_range[0], t_range[1]+dt, dt)
    n_vector = np.empty(len(t))
    current_vector = np.empty([len(current_funcs),len(t)])
    c_vector = n_vector.copy()
    
    volt_vector = np.zeros(len(t))
    volt_vector[0] = volt_init
    #pdb.set_trace()
    n_vector[0] = n_init
    c_vector[0] = c_init
    
    for i in np.arange(len(t)-1):
        tempVoltage = volt_vector[i]
        
        #compute c and n at current time step
        n_vector[i+1] = n_vector[i] + dt * n_func(n_vector[i],tempVoltage,t[i])
        c_vector[i+1] = c_vector[i] + dt * c_func(c_vector[i],tempVoltage,t[i])
        
        # compute I_Ca at current time step
        current_vector[0,i] = current_funcs[0](tempVoltage)
        
        # compute I_K_HH at current time step
        current_vector[1,i] = current_funcs[1](n_vector[i],tempVoltage)
        
        # compute I_K_Ca at current time step
        current_vector[2,i] = current_funcs[2](c_vector[i],tempVoltage)
        
        # compute I_L at current time step
        current_vector[3,i] = current_funcs[3](tempVoltage)
        
        volt_vector[i+1] = volt_vector[i] + dt * volt_func(current_vector[0,i],current_vector[1,i],current_vector[2,i],current_vector[3,i])                
        
    return t, n_vector, c_vector, current_vector, volt_vector

def n_ode(n, V, t):
    a = a_n(V)
    b = b_n(V)
    result = a * (1 - n) - b * n
    return result


def concentration_ode(c, V, t):
    f = 0.007
    k1 = 0.0275 #microM*cm^2/nC
    kc = 0.02 #ms^-1
    return f * (-k1*I_Ca(V) - c*kc)

#We changed the values accordingly
#V + V* = V - 15 for m and h
#V + V* = V - 35 for n

def a_n(V):
    V = V - 35.0
#    return (0.01* (V + 20)) / (1 - np.exp(-0.1 * (V + 20)))
#    return (0.01 * (V - 10.0))/(np.exp((V-10.0)/10.0) - 1.0)
    return ((V+55.0)/100.0)/(1.0 - np.exp(-(V + 55.0)/10.0))
def b_n(V):
    V = V - 35.0
#    return 0.125 * np.exp(-0.0125 *(V + 20))
#    return 0.125 * np.exp(V/80.0)
    return np.exp(-(V + 65.0)/80.0)/8.0

def a_m(V):
    V = V-15.0
#    return 0.1* (V + 25) /(1 - np.exp(-0.1 *(V + 25)))
#    return (0.1 * (V - 25.0)) / ((np.exp((V-25.0)/10.0)) - 1.0)
    return ((V+40.0)/10.0)/(1.0 - np.exp(-(V + 40.0)/10.0))
    
def a_h(V):
    V = V-15.0
#    return 0.07 * np.exp(-0.05 *(V + 50))
#    return 0.07 * np.exp(V/20.0)
    return 7.0*np.exp(-(V + 65.0)/20.0)/100.0

def b_m(V):
    V = V-15.0
#    return 4.0 * np.exp(-0.0556 *(V + 50))
#    return 4.0*np.exp(V/18.0)
    return 4.0*np.exp(-(V + 65.0)/18.0)

def b_h(V):
    V = V-15.0
#    return 1.0 /(1+ np.exp(-0.1 *(V+20)))
#    return (np.exp((V-30.0)/10.0)+1)**(-1)
    return 1.0/(1.0 + np.exp(-(V + 35.0)/10.0))

def m_steady(V):
    return a_m(V)/(a_m(V) + b_m(V))

def h_steady(V):
    return a_h(V)/(a_h(V) + b_h(V))


def voltage_ode (I_Ca, I_K_HH, I_K_Ca, I_L):
    C_m = 1
    return (-I_Ca - I_K_HH - I_K_Ca - I_L)/C_m

def I_K_HH(n,V):
    g_K = 5.5 #mS/cm^2
    V_K = -75 # mV
    return g_K* n**4 * (V-V_K)

def I_K_Ca(c, V):
    g_K_Ca = 0.05 #mS/cm^2
    K_d = 1 # micro mol per liter
    V_K = -75 #mV
    return c * g_K_Ca / (K_d + c) * (V - V_K)

def I_L(V):
    g_L = 0.03 #mS/cm^2
    V_L = -40 #mV
    return g_L* (V-V_L)

def I_Ca(V):
    g_Ca = 8 #mS/cm^2
    V_Ca = 100 #mV
    return g_Ca * m_steady(V)**3 * h_steady(V) * (V - V_Ca)
```

```python
def compute_fr(volt,threshold):
    volt_below_thr = volt<threshold
    volt_above_thr = volt>threshold
    spike_idc = np.logical_and(volt_below_thr, np.roll(volt_above_thr,-1))
    nspikes = np.sum(spike_idc)
    return nspikes, spike_idc

nspikes, spike_idc = compute_fr(volt,-20)
print spike_idc

nspikes, spike_idc = compute_fr(volt,-20)
plt.figure()
timeWindow = np.arange(int(time_in_seconds*1000/dt))
plt.plot(spike_idc)
```
